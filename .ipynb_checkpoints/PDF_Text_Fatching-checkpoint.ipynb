```python
import camelot
import pandas as pd
import numpy as np

pdf_path = "Ingrediants_with_it\'s_usage.pdf"

tables = camelot.read_pdf(pdf_path, flavor="stream", pages="9-520")
```

```python
ingrediants_df = pd.DataFrame()
for i in range(tables.n):
    ingrediants_df = pd.concat([ingrediants_df, tables[i].df[3:]], ignore_index=True)
```

```python
ingrediants_df.isnull().sum()
```
