```python
def initializeDriver():
    path="D:/Selenium/chromedriver_win32/chromedriver.exe"
    driver = webdriver.Chrome(path)
    return driver

def closePopUp():
    try:
        popup = WebDriverWait(driver, 10).until(
                EC.visibility_of_element_located((By.CLASS_NAME, "UpdateCityModal__modal-body___H98MG"))
        )
        cancel_button = popup.find_element(By.CLASS_NAME, "UpdateCityModal__cancel-btn___2jWwS")
        cancel_button.click()
    except:
        print("Pop Up not found!!")

def getLinks(products):
    productLinks=[]
    for element in products:
        tags = element.find_elements_by_tag_name('a')
        for tag in tags:
            link = tag.get_attribute('href')
            productLinks.append(link)
    return productLinks

def findAndClickNext():
    next_button = driver.find_elements_by_class_name("next")
    if next_button:
        next_btn = WebDriverWait(driver, 5).until(
                    EC.visibility_of_element_located((By.CLASS_NAME, "next"))
            )
        next_button = next_btn.find_element(By.CLASS_NAME, "button-text")
        next_button.click()
        return True
    else: 
        return False

def findingProducts(link):
    driver.get(link)
    closePopUp()
    products = driver.find_elements_by_class_name("style__product-box___liepi")
    return products
```

```python
driver = initializeDriver()
```

```python
for link in links:
    all_product_links = []
    first_pages_value = link['pages']
    for j in range(1, first_pages_value + 1):
        url = str(link['link']) + str(j)
        print(url)
        products = findingProducts(url)
        product_links = getLinks(products)
        all_product_links.extend(product_links)
        
        next_button_exists= findAndClickNext()

        if not next_button_exists:
            break

        df = pd.DataFrame(all_product_links)
        df.to_csv(link['category']+".csv")
```
