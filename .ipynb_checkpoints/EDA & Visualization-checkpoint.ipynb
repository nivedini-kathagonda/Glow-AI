```python
category_counts = df['category'].value_counts()

plt.figure(figsize=(20,10))
category_counts.plot(kind='barh', color='salmon')
plt.title('Count plot of Categories')
plt.xlabel('Number of Values')
plt.ylabel('Product Category')
plt.grid(axis='x')
plt.savefig("Images_Lab/Count plot of Categories")
```

### Observation:

Distribution of products across different categories:

- hair-skin-nails: 3,085 products
- facewash: 2,370 products
- face-serum-oil: 1,157 products
- face-scrubs-exfoliators: 948 products
- facial kit - sheet mask: 574 products

```python
plt.figure(figsize=(20,10))
plt.hist(df['rating_element'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Product Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.grid(axis='y')
plt.savefig("Images_Lab/Distribution of Product Ratings")
```

### Observation:
- A large number of products receive ratings close to 5 (high satisfaction).
- The bulk of goods have ratings ranging from 4 to 5.
- Fewer products get ratings lower than 4.

```python
df['price_cleaned'] = df['price'].str.replace('₹', '').str.replace(',', '').astype(float)

avg_price_per_category = df.groupby('category')['price_cleaned'].mean().sort_values(ascending=False)

plt.figure(figsize=(20,10))
avg_price_per_category.plot(kind='barh', color='salmon')
plt.title('Average Price of Products Across Different Categories')
plt.xlabel('Average Price')
plt.ylabel('Product Category')
plt.grid(axis='x')
plt.savefig("Images_Lab/Average Price of Products Across Different Categories")
```

### Observation:

- On average, face-serum-oil is the most expensive category.
- Baby sunscreens and skin pigmentation products are considerably more expensive on average.
- Lip balm, makeup removal wipes, and toners-astringents are some of the more cheap categories.

```python
avg_rating_per_category = df.groupby('category')['rating_element'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,8))
avg_rating_per_category.plot(kind='barh', color='lightgreen')
plt.title('Average Ratings of Products Across Different Categories')
plt.xlabel('Average Rating')
plt.ylabel('Product Category')
plt.xlim(0, 5)
plt.grid(axis='x')
plt.savefig("Images_Lab/Average Ratings of Products Across Different Categories")
```

### Observation:
- Several product categories, such as baby sunscreens and skin pigmentation, have average ratings near 5 stars.
- Toners-astringents, facewash, and face scrubs-exfoliators have slightly lower average ratings than others, but they are still rather high (above 4).


```python
stop_words = set(stopwords.words('english'))
all_descriptions = ' '.join(df['description'].dropna())
wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      stopwords = stop_words,
                      min_font_size = 10).generate(all_descriptions)

plt.figure(figsize=(8,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.title('Word Cloud for Product Descriptions')
plt.savefig("Images_Lab/Word Cloud for Product Description")
```

```python
description_tokens = ' '.join(df['description'].dropna()).lower().split()

description_filtered_tokens = [word for word in description_tokens if word not in stop_words and word.isalnum()]

description_unigrams = description_filtered_tokens
description_bigrams = list(nltk.bigrams(description_filtered_tokens))
description_trigrams = list(nltk.trigrams(description_filtered_tokens))

description_unigram_counts = Counter(description_unigrams).most_common(10)
description_bigram_counts = Counter(description_bigrams).most_common(10)
description_trigram_counts = Counter(description_trigrams).most_common(10)

description_unigram_counts, description_bigram_counts, description_trigram_counts
```

```python
# Extracting data for visualization
bigram_labels = [" ".join(bigram[0]) for bigram in description_bigram_counts]
bigram_values = [bigram[1] for bigram in description_bigram_counts]

trigram_labels = [" ".join(trigram[0]) for trigram in description_trigram_counts]
trigram_values = [trigram[1] for trigram in description_trigram_counts]

# Plotting bigrams and trigrams
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Bigrams
ax[0].barh(bigram_labels, bigram_values, color='lightblue')
ax[0].set_title('Top 10 Bigrams in Product Descriptions')
ax[0].set_xlabel('Frequency')
ax[0].set_ylabel('Bigrams')
ax[0].invert_yaxis()

# Trigrams
ax[1].barh(trigram_labels, trigram_values, color='lightcoral')
ax[1].set_title('Top 10 Trigrams in Product Descriptions')
ax[1].set_xlabel('Frequency')
ax[1].set_ylabel('Trigrams')
ax[1].invert_yaxis()

plt.tight_layout()
plt.savefig("Images_Lab/Bigram-Trigram")
```
